**Question 1: What is hypothesis testing in statistics?**

Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population using sample data.

Simple Explanation

Hypothesis testing helps you check whether a claim (statement) about a population is likely to be true based on sample evidence.

Key Concepts

Null Hypothesis (H₀)
A default assumption or claim.
Example: “The average height of students = 160 cm.”

Alternative Hypothesis (H₁ or Ha)
What you want to test or prove.
Example: “The average height of students is not 160 cm.”

Test Statistic
A value calculated from sample data (like z-score or t-score) used to test the claim.

p-value
The probability of observing the sample result if the null hypothesis were true.

Low p-value (< 0.05): Reject H₀

High p-value (> 0.05): Fail to reject H₀

Decision
Based on the p-value, you decide whether the evidence is strong enough against the null hypothesis.

In One Line

Hypothesis testing helps decide whether sample data supports or contradicts a claim about a population.

**Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?**


Null Hypothesis (H₀)

The null hypothesis is a statement that says there is no effect, no difference, or no relationship in the population.
It represents the default or baseline assumption.

Examples:

A new medicine has no effect compared to the old one.

The average income of two groups is the same.

A teaching method does not change student performance.

Alternative Hypothesis (H₁ or Ha)

The alternative hypothesis is a statement that says there IS an effect, a difference, or a relationship.
It is what the researcher wants to prove.

Examples:

A new medicine does affect patient recovery.

The average income of two groups is different.

A teaching method improves student performance.


| Feature     | Null Hypothesis (H₀)      | Alternative Hypothesis (H₁)          |
| ----------- | ------------------------- | ------------------------------------ |
| Meaning     | No effect / No difference | There is an effect / difference      |
| Purpose     | Default assumption        | What the researcher wants to support |
| Test Result | Rejected or Not Rejected  | Supported if H₀ is rejected          |
| Example     | Mean = 50                 | Mean ≠ 50 (or > 50, < 50)            |


Summary

H₀: "Nothing is happening."

H₁: "Something is happening."









**Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test**.


Significance Level (α) in Hypothesis Testing

The significance level, usually denoted by α (alpha), is the probability of rejecting the null hypothesis (H₀) when it is actually true.
This is also known as the Type I error rate.

Common values for α are 0.05, 0.01, or 0.10.

Why is the significance level important?

It sets the threshold for decision-making in hypothesis testing.

After calculating the p-value, we compare it with α:

If p-value ≤ α:
Reject the null hypothesis (evidence is strong enough).

If p-value > α:
Fail to reject the null hypothesis (not enough evidence).


| Scenario        | Decision          | Meaning                                                        |
| --------------- | ----------------- | -------------------------------------------------------------- |
| **p-value ≤ α** | Reject H₀         | Data provides significant evidence against the null hypothesis |
| **p-value > α** | Fail to reject H₀ | No strong evidence to say the null hypothesis is false         |


Example

If α = 0.05, you are allowing a 5% chance of incorrectly rejecting the null hypothesis.

It means:

“I am 95% confident that the difference or effect is real, not due to random chance.”

In summary

Significance level α controls the risk of false positives.

It determines the cut-off for deciding whether the evidence is strong enough to reject H₀.

Lower α = more strict test (less chance of Type I error).




**Question 4: What are Type I and Type II errors? Give examples of each.**

Type I Error (False Positive)

A Type I error occurs when you reject a true null hypothesis.
In simple words: You think something is happening when actually it is not.

Example: Medical Test

A test shows a patient has a disease, but they actually do NOT have it.
This is a false alarm.

Other Examples

A smoke alarm rings, but there is no fire.

A court declares an innocent person guilty.


Type II Error (False Negative)

A Type II error occurs when you fail to reject a false null hypothesis.
Meaning: You think nothing is happening when actually it IS happening.

Example: Medical Test

A test shows a patient does NOT have a disease, but they actually DO have it.
This is missing something important.

Other Examples

A smoke alarm does NOT ring, even though there is a fire.

A court sets a guilty person free.


| Error Type  | What Happens?           | Simple Meaning | Example                         |
| ----------- | ----------------------- | -------------- | ------------------------------- |
| **Type I**  | Reject true H₀          | False Positive | Saying a healthy person is sick |
| **Type II** | Fail to reject false H₀ | False Negative | Saying a sick person is healthy |


**Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.**


| **Z-Test**                                                       | **T-Test**                                                                     |
| ---------------------------------------------------------------- | ------------------------------------------------------------------------------ |
| Used when the **population standard deviation (σ)** is **known** | Used when the **population standard deviation is NOT known**                   |
| Typically used for **large sample sizes (n ≥ 30)**               | Typically used for **small sample sizes (n < 30)**                             |
| Uses the **Z distribution (normal distribution)**                | Uses the **t-distribution** (which is similar to normal but has thicker tails) |
| More accurate when sample size is large                          | More accurate when sample size is small and σ is unknown                       |


When to Use Each
Use a Z-Test when:

The population standard deviation (σ) is known.

The sample size is large (n ≥ 30).

Data follows (or approximately follows) a normal distribution.
Examples: Testing whether the average height of adults is 170 cm when population variance is known.

**Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.**
(Include your Python code and output in the code box below.)

```python
# Below is a complete Python script that:

# Simulates data from a normal distribution

# Calculates the 95% confidence interval for the mean

# Plots the data using Matplotlib

# Prints the CI and mean

# You can copy-paste this directly into your code box or VS Code.


import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 1. Simulate data from a normal distribution
np.random.seed(42)                   # for reproducibility
data = np.random.normal(loc=50,      # mean
                        scale=10,    # standard deviation
                        size=1000)   # sample size

# 2. Calculate statistics
sample_mean = np.mean(data)
sample_std  = np.std(data, ddof=1)
n = len(data)

# 95% confidence interval for mean
confidence_level = 0.95
alpha = 1 - confidence_level
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)

margin_of_error = t_critical * (sample_std / np.sqrt(n))
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print("Sample Mean:", sample_mean)
print("95% Confidence Interval for Mean:")
print(f"({lower_bound:.4f}, {upper_bound:.4f})")

# 3. Plot the data
plt.figure(figsize=(10, 5))
plt.hist(data, bins=30, edgecolor='black')
plt.title("Histogram of Simulated Normal Data")
plt.xlabel("Value")
plt.ylabel("Frequency")

# add a vertical line for the sample mean
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {sample_mean:.2f}")
plt.legend()
plt.show()

# What the script does

# Generates 1000 values from N(50, 10^2)

# Computes sample mean & standard deviation

# Uses the t-distribution to compute the 95% confidence interval

# Shows a histogram with the mean highlighted
```

Question 9: Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.
(Include your Python code and output in the code box below.)

```python
What Z-scores Represent

A Z-score tells you how many standard deviations a data point is from the mean.

Z = 0 → value is exactly at the mean

Z = +1 → value is 1 standard deviation above the mean

Z = −2 → value is 2 standard deviations below the mean

Z-scores help in standardizing data, detecting outliers, and comparing data across different scales.

✅ Python Code (Function + Histogram)
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Z-scores
def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std
    return z_scores

# Sample dataset
data = [7, 8, 5, 9, 6, 7, 8, 9, 4, 6, 7, 8]

# Calculate Z-scores
z_scores = calculate_z_scores(np.array(data))
print("Z-scores:\n", z_scores)

# Plot histogram of standardized data
plt.hist(z_scores)
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.title("Histogram of Standardized Data (Z-scores)")
plt.show()

📌 Output Explanation

The code prints an array of Z-scores.

The histogram shows how the standardized values are distributed around 0, which is the mean in standard normal units.
```
